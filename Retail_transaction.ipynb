```python
import pandas as pd
```

```python
dept = pd.read_csv(r"E:\2026\PROJECCTS\Instakart_REtail_project\departments.csv")
```

```python
prior_order_prod = pd.read_csv(r"E:\2026\PROJECCTS\Instakart_REtail_project\order_products__prior.csv")
```

```python
orders= pd.read_csv(r"E:\2026\PROJECCTS\Instakart_REtail_project\orders.csv")
```

```python
prod = pd.read_csv(r"E:\2026\PROJECCTS\Instakart_REtail_project\products.csv")
```

```python
print(f" columns of department {dept.columns}")
print(f"coulmns of order_products_prior table {prior_order_prod.columns}")
print(f"columns of orders table {orders.columns}")
print(f"columns of products table {prod.columns}")
```

```python
first_half_merge = pd.merge(prod,dept, on="department_id", how = "left")
first_half_merge
```

```python
half_merged =  pd.merge(orders, prior_order_prod, on="order_id")
half_merged
```

```python
sales_lines_clean = pd.merge(half_merged, first_half_merge, on = "product_id")
sales_lines_clean.head()
```

```python
sales_lines_clean = sales_lines_clean.drop(["eval_set","order_number","add_to_cart_order","reordered","aisle_id"],axis=1)
sales_lines_clean.head()
```

```python
sales_lines_clean.shape
```

```python
sales_lines_clean["product_name"] = sales_lines_clean["product_name"].astype(str).str.strip().str.lower()
sales_lines_clean["department"]=sales_lines_clean["department"].astype(str).str.strip().str.lower()
sales_lines_clean.head()
```

```python
sales_lines_clean["days_since_prior_order"]=sales_lines_clean["days_since_prior_order"].fillna(0)
sales_lines_clean.head()
```

```python
sales_lines_clean[["order_id","user_id","order_dow","order_hour_of_day","days_since_prior_order","product_id","department_id"]]=sales_lines_clean.loc[0:][["order_id","user_id","order_dow","order_hour_of_day","days_since_prior_order","product_id","department_id"]].astype(int)
```

```python
import random
product_price = {}
price = []
unique=sales_lines_clean.iloc[0:,6].unique()
for i in range(len(unique)):
    price.append(random.randint(1,15))
product_price = dict(zip(unique, price))
print(dict(list(product_price.items())[:7]))


#print(unique_values)
```

```python
sales_lines_clean["price"] = sales_lines_clean["product_name"].map(product_price)
sales_lines_clean.head()
```

```python
import datetime 
#print(dir(datetime.datetime))
start_date = datetime.date(2025,1,6)
date =[]

for i in range (len(sales_lines_clean)):
    rangedate = datetime.timedelta(random.randint(0,41))
    date.append(start_date + rangedate)
sales_lines_clean["date"] = date
```

```python
sales_lines_clean["date"] = pd.to_datetime(sales_lines_clean["date"])
diff = sales_lines_clean["date"] - pd.Timestamp(start_date)
diff=diff.dt.days
sales_lines_clean["week"] = (diff//7)+1
sales_lines_clean.head(10)
```

```python
sales_lines_clean=sales_lines_clean.sort_values(by=["product_id","date"], ascending= True)
```

```python
sample_data = sales_lines_clean.sample(100000, random_state=25)
```

```python
qty =[]
for i in range(len(sample_data)):
    qty.append(random.randint(1,6))
sample_data["qty"]=qty
```

```python
sample_data["row_revenue"] = sample_data["price"] * sample_data["qty"]
```

```python
day_number= sample_data["date"] - pd.Timestamp("2025-01-06")
day_number= day_number.dt.days
sample_data["day_number"]= day_number+1
```

```python
sample_data=sample_data.sort_values(by=["date","product_id"],ascending=True)
```

```python
#Top 10 products by quantity sold Include: product name, department, total qty, total revenue
sellers = sample_data.groupby(['week','product_name','department']).agg(total_qty=('qty','sum'), total_revenue=('row_revenue','sum'))
sellers= sellers.sort_values(by =['week','total_qty'], ascending=[True, False])
weekly_best_sellers=sellers.groupby('week').head(10)
weekly_best_sellers.to_csv("weekly_best_sellers.csv")
```

```python
#Bottom 5 products by quantity sold Include: product name, department, total qty, total revenue
sellers = sample_data.groupby(['week','product_name','department']).agg(total_qty=('qty','sum'), total_revenue=('row_revenue','sum'))
sellers= sellers.sort_values(by=['week','total_qty'], ascending=[True,True])
weekly_worst_sellers = sellers.groupby('week').head(5)
weekly_worst_sellers.to_csv("weekly_worst_sellers.csv")
```

```python
#otal sales per department Include: week,department, total qty, total revenue, number of different products sold

sales= sample_data.groupby(['week','department']).agg(total_qty=('qty','sum'), total_revenue=('row_revenue','sum'),num_products=("product_name","nunique"))
weekly_category_sales= sales
weekly_category_sales.to_csv("weekly_category_sales.csv")
```
